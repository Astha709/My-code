```python
import pandas as pd

path = "/content/drive/MyDrive/CSV File/training.1600000.processed.noemoticon.csv"
df = pd.read_csv(path)

print("Clean data loaded")
print(df.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout

```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

MAX_VOCAB = 20000
MAX_LEN = 50  # tweets are short

tokenizer = Tokenizer(num_words=MAX_VOCAB, oov_token="")
tokenizer.fit_on_texts(df['clean_text'])

X = tokenizer.texts_to_sequences(df['clean_text'])
X = pad_sequences(X, maxlen=MAX_LEN, padding='post')

y = df['target'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42, stratify=y)
```

```python
model = Sequential([
    Embedding(input_dim=MAX_VOCAB, output_dim=128, input_length=MAX_LEN),
    LSTM(128, dropout=0.2, recurrent_dropout=0.2),
    Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=3,
    batch_size=512
)

```

```python
model.evaluate(X_test, y_test)

model.save("/content/drive/MyDrive/CSV File/twitter_sentiment_project/models/lstm_sentiment.h5")
print("✅ LSTM model saved")
```
